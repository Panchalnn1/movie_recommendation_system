```python
movies= movies.drop('keywords_y', axis=1)
movies.head()
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords_x'] = movies['keywords_x'].apply(convert)
```

```python

movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def find_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies.head()
```

```python
def marge(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(marge)
movies['crew'] = movies['crew'].apply(marge)
movies['genres'] = movies['genres'].apply(marge)
movies['keywords_x'] = movies['keywords_x'].apply(marge)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['info'] = movies['overview'] + movies['genres'] + movies['keywords_x'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords_x','cast','crew'])
```

```python
new.head()
```

```python
new['info'] = new['info'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=300,stop_words='english')
```

```python
vector = cv.fit_transform(new['info']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    recommended_movies = "\n".join([new.iloc[i[0]].title for i in distances[1:6]])
    return recommended_movies
```

```python
print(recommend('Jumanji'))
```

```python
import gradio as gr
import pandas as pd

def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    recommended_movies = "\n".join([new.iloc[i[0]].title for i in distances[1:6]])
    return recommended_movies

# Creating the Gradio interface
interface = gr.Interface(
    fn=recommend,
    inputs="text",
    outputs="text",
    title="Movie Recommender",
    description="Enter a movie title to get recommendations"
)

# Launch the interface
interface.launch()
```
